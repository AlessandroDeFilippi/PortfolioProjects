```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica il file CSV con separatore ';'
df = pd.read_csv('popolazione.csv', sep=';')
df.columns = df.columns.str.strip()

# Calcola il totale residenti per anno e cittadinanza (italiani vs non italiani)
df['Categoria'] = df['Cittadinanza'].apply(lambda x: 'Italiani' if x == 'Italia' else 'Stranieri')

# Raggruppa per anno e categoria, sommando i residenti
df_totale = df.groupby(['Anno', 'Categoria'])['Residenti'].sum().unstack()

# Visualizza il grafico
df_totale.plot(kind='line', marker='o', figsize=(10,6),
               title='Residenti Italiani e Stranieri nel tempo')
plt.ylabel('Numero Residenti')
plt.xlabel('Anno')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# 1. Carica il file con separatore ';'
df = pd.read_csv('popolazione.csv', sep=';')
df.columns = df.columns.str.strip()

# 2. Trasforma le et da stringa a intero (assicurati che siano numeri validi)
df['Eta'] = pd.to_numeric(df['Eta'], errors='coerce')

# 3. Rimuovi righe senza et valida
df = df.dropna(subset=['Eta'])

# 4. Crea le fasce d'et personalizzate
def fascia_eta(et):
    if et <= 14:
        return '014 (Infanzia e adolescenza)'
    elif et <= 24:
        return '1524 (Giovani)'
    elif et <= 44:
        return '2544 (Giovani adulti e famiglie)'
    elif et <= 64:
        return '4564 (Adulti maturi)'
    elif et <= 84:
        return '6584 (Anziani attivi)'
    else:
        return '85+ (Grandi anziani)'

df['FasciaEta'] = df['Eta'].apply(fascia_eta)

# 5. Crea categoria Italiani/Stranieri
df['Nazionalit'] = df['Cittadinanza'].apply(lambda x: 'Italiana' if x.strip() == 'Italia' else 'Straniera')

# 6. Raggruppa per Anno, Quartiere, FasciaEt, Nazionalit e somma i residenti
df_aggregato = df.groupby(['Anno', 'Nil', 'FasciaEta', 'Nazionalit'])['Residenti'].sum().reset_index()

# 7. Salva il file aggregato per Tableau
df_aggregato.to_csv('popolazione_aggregata_per_tableau.csv', index=False)

print("File salvato: popolazione_aggregata_per_tableau.csv")
```

```python
import pandas as pd

# === 1. Carica il file CSV originale ===
# Assumiamo che il separatore sia il punto e virgola
df = pd.read_csv('popolazione.csv', sep=';')
df.columns = df.columns.str.strip()  # Rimuove spazi indesiderati nei nomi delle colonne

# === 2. Pulisci e converte la colonna 'Eta' ===
df['Eta'] = pd.to_numeric(df['Eta'], errors='coerce')  # Converte a numero, gestisce errori
df = df.dropna(subset=['Eta'])                         # Rimuove righe con et non valida
df['Eta'] = df['Eta'].astype(int)                      # Converte definitivamente in int

# === 3. Crea le fasce d'et semplificate ===
def fascia_eta(eta):
    if eta <= 14:
        return '0-14 (Infanzia e adolescenza)'
    elif eta <= 24:
        return '15-24 (Giovani)'
    elif eta <= 44:
        return '25-44 (Giovani adulti e famiglie)'
    elif eta <= 64:
        return '45-64 (Adulti maturi)'
    elif eta <= 84:
        return '65-84 (Anziani attivi)'
    else:
        return '85+ (Grandi anziani)'

df['FasciaEta'] = df['Eta'].apply(fascia_eta)

# === 4. Classifica le cittadinanze: Italiana o Straniera ===
df['Nazionalit'] = df['Cittadinanza'].apply(lambda x: 'Italiana' if str(x).strip() == 'Italia' else 'Straniera')

# === 5. Raggruppa i dati per Anno, Quartiere, Fascia d'et, Nazionalit ===
df_aggregato = df.groupby(
    ['Anno', 'Nil', 'FasciaEta', 'Nazionalit']
)['Residenti'].sum().reset_index()

# === 6. Esporta in CSV compatibile con Excel e Tableau ===
df_aggregato.to_csv('popolazione_aggregata_per_tableau.csv', index=False, encoding='utf-8-sig')

print("File generato con successo: popolazione_aggregata_per_tableau.csv")
```
